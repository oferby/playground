```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.array(range(20))
x
```

```python
y = np.array(range(20))
y
```

```python
x, y = np.meshgrid(x, y)
z = x + y
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
ax.plot_surface(x,y,z)
plt.show()
```
