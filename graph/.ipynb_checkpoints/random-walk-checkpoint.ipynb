# Random Walk

```python
def init_walks_property(G):
    for e in list(G.edges): G.edges[e]['walks'] = 0
```

```python
def init_graph():
    DG = nx.DiGraph()
    DG.add_edges_from([(1,3,{'walks':0}),(2,4,{'walks':0}),(2,5,{'walks':0}),(4,3,{'walks':0}),(5,4,{'walks':0}),(5,6,{'walks':0}),(5,7,{'walks':0}),(3,8,{'walks':0}),(8,7,{'walks':0})])
    return DG
```

```python
def init_tree():
    DG = nx.DiGraph()
    DG.add_edges_from([(1,2),(1,3),(2,4),(2,5),(3,6),(3,7)])
    init_walks_property(DG)
    return DG
```

```python
def get_random_next(x):
    d = DG.out_degree(x)
    if d == 0:
        return None
    i = r.randint(0,d-1)
    return list(DG.successors(x))[i]
```

```python
def get_starting_node():
    nodes = list(DG.nodes)
    i = r.randint(0, len(nodes) - 1)
    return nodes[i]
```

```python
def get_limited_starting_node():
    nodes = []
    for n in DG.nodes:
        if DG.in_degree(n) == 0:
            nodes.append(n)
    i = r.randint(0, len(nodes) - 1)
    return nodes[i]    
```

```python
def do_random_walk(k, limited = False, endless = False):
    for j in range(k):
        i = 0
        if limited:
            a = get_limited_starting_node()
        else:
            a = get_starting_node()
        while True:
            i+=1
            if endless:
                if i >= 10:
                    break
            b = get_random_next(a)
            if b is None:
                break
            DG.edges[a,b]['walks']+=1
            a = b
```

```python
def print_walks(with_probability=True):
    
    dgl = list(DG.edges)
    total = 0
    if with_probability:
        for i in range(len(dgl)):
            total += DG.edges[dgl[i]]['walks']
    
    for i in range(len(dgl)):
        if with_probability:
            print(dgl[i], DG.edges[dgl[i]]['walks'], "{:.2f}%".format(100 * DG.edges[dgl[i]]['walks'] / total))
        else:
            print(dgl[i], DG.edges[dgl[i]]['walks'])
```

## Probability Matrix

```python
def get_degree_matrix(M):
    x,_ = M.shape
    u = np.ones(x)
    d = M.dot(u)
    D = d * np.eye(x)
    return D
```

```python
def get_probability_matrix(M):
    x,_ = M.shape
    u = np.ones(x)
    d = M.dot(u)
    
    for i in range(len(d)):
        if d[i] != 0:
            d[i] = 1 / d[i]
    
    D = d * np.eye(x)
    return D.dot(M)
```

```python
def get_adj_matrix(DG):
    AM = nx.linalg.graphmatrix.adjacency_matrix(DG)
    M = scipy.sparse.csr_matrix.toarray(AM)
    return M
```

```python
def get_probability_after_time(M, p, t):
    PM = get_probability_matrix(M)

    tp = np.copy(p)
    
    for i in range(t):
        p = p.dot(PM)
        p = p / sum(p)
        tp += p
    
    tp = tp / sum(tp)
    return p, tp    
```
